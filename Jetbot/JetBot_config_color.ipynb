```python
import cv2
import numpy as np
import ipywidgets.widgets as widgets
from IPython.display import display
from jetbot import Camera, bgr8_to_jpeg
import traitlets
import os

# 1. SAFE CAMERA CONNECTION
try:
    camera = Camera.instance()
    camera.unobserve_all() 
    print("âœ… Camera connected and state reset.")
except Exception as e:
    print(f"Camera Error: {e}")

# 2. LOAD PREVIOUS CONFIG (IF EXISTS)
# Default values
def_h_min, def_s_min, def_v_min = 20, 80, 80
def_h_max, def_s_max, def_v_max = 45, 255, 255

if os.path.exists('yellow_config.npy'):
    try:
        conf = np.load('yellow_config.npy')
        def_h_min, def_s_min, def_v_min = conf[0], conf[1], conf[2]
        def_h_max, def_s_max, def_v_max = conf[3], conf[4], conf[5]
        print("ðŸ“‚ Found previous configuration file -> Loaded into sliders.")
    except:
        print("âš ï¸ Configuration file error -> Using defaults.")
else:
    print("â„¹ï¸ No configuration file found -> Using defaults.")

# 3. INTERFACE (UI)
image_widget_y = widgets.Image(format='jpeg', width=300, height=300)
mask_widget_y = widgets.Image(format='jpeg', width=300, height=300)

# Create Sliders with initial values loaded from file
h_min_y = widgets.IntSlider(min=0, max=179, value=def_h_min, description='H Min')
s_min_y = widgets.IntSlider(min=0, max=255, value=def_s_min, description='S Min')
v_min_y = widgets.IntSlider(min=0, max=255, value=def_v_min, description='V Min')
h_max_y = widgets.IntSlider(min=0, max=179, value=def_h_max, description='H Max')
s_max_y = widgets.IntSlider(min=0, max=255, value=def_s_max, description='S Max')
v_max_y = widgets.IntSlider(min=0, max=255, value=def_v_max, description='V Max')

save_button_y = widgets.Button(description='Save Yellow Config', button_style='success')
output_y = widgets.Output()

def save_yellow_config(b):
    config = np.array([
        h_min_y.value, s_min_y.value, v_min_y.value,
        h_max_y.value, s_max_y.value, v_max_y.value
    ], dtype=np.int32) # Note: integer data type
    np.save('yellow_config.npy', config)
    with output_y:
        output_y.clear_output()
        print(f"âœ… SAVED Yellow parameters: {config}")

save_button_y.on_click(save_yellow_config)

print("--- STEP 1: CALIBRATE YELLOW LINE COLOR ---")
display(widgets.VBox([
    widgets.HBox([image_widget_y, mask_widget_y]),
    widgets.HBox([h_min_y, s_min_y, v_min_y]),
    widgets.HBox([h_max_y, s_max_y, v_max_y]),
    save_button_y,
    output_y
]))

# 4. IMAGE PROCESSING
def update_yellow(change):
    new_image = change['new']
    hsv = cv2.cvtColor(new_image, cv2.COLOR_BGR2HSV)
    lower = np.array([h_min_y.value, s_min_y.value, v_min_y.value])
    upper = np.array([h_max_y.value, s_max_y.value, v_max_y.value])
    mask = cv2.inRange(hsv, lower, upper)
    
    image_widget_y.value = bgr8_to_jpeg(new_image)
    ret, jpeg = cv2.imencode('.jpg', mask)
    mask_widget_y.value = jpeg.tobytes()

try:
    camera.observe(update_yellow, names='value')
except Exception as e:
    print("Error connecting image stream.")
```
